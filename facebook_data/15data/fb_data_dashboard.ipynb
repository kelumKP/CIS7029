```python
import pathlib
import holoviews as hv
import hvplot.pandas # noqa
import numpy as np
import pandas as pd
import panel as pn
import datetime as dt
import calendar

pn.extension(sizing_mode='stretch_width')
```

```python
files = ['PrimeLands.xlsx', 'BlueOcean.xlsx', 'CBHLands.xlsx', 'HomeLands.xlsx','HouseAndLandsLk.xlsx','SkyLine.xlsx']
```

```python
df = pd.DataFrame()
for file in files:
     if file.endswith('.xlsx'):
         df = df.append(pd.read_excel(file, sheet_name="posts"), ignore_index=True) 
df.head()
```

```python
df.columns
```

```python
df[df['company'] == 'Prime Lands']
```

```python
df['post_id'] = df['post_id'].replace('', np.nan)
df = df.dropna(axis=0, subset=['post_id'])

df['created_time'] = df['created_time'].replace('', np.nan)
df = df.dropna(axis=0, subset=['created_time'])

df['updated_time'] = df['updated_time'].replace('', np.nan)
df = df.dropna(axis=0, subset=['updated_time'])

df
```

```python
df['created_date'] = pd.to_datetime(df['created_time']).dt.date
df
```

```python
# Fill NAs with 0s and create GDP per capita column
df = df.fillna(0)
```

```python
df['created_year'] = pd.to_datetime(df['created_time']).dt.year
df['created_month'] = pd.to_datetime(df['created_time']).dt.month

df
```

```python
df['shares'] = df['shares_count'].replace(np.nan, 0)
df['comments'] = df['comments_count'].replace(np.nan, 0)
df['reactions'] = df['reactions_count'].replace(np.nan, 0)
df['like'] = df['like_count'].replace(np.nan, 0)
df['love'] = df['love_count'].replace(np.nan, 0)
df['haha'] = df['haha_count'].replace(np.nan, 0)
df['wow'] = df['wow_count'].replace(np.nan, 0)
df['sad'] = df['sad_count'].replace(np.nan, 0)
df['angry'] = df['angry_count'].replace(np.nan, 0)

df['created_time'] = pd.to_datetime(df['created_time'])
df['updated_time'] = pd.to_datetime(df['updated_time'])

df.sort_values(by='created_time', inplace=True)

df
```

```python
# Make DataFrame Pipeline Interactive
idf = df.interactive()

df.set_index('created_date', inplace=True)
```

```python
months_slider = pn.widgets.IntSlider(name='Month', start=1, end=12, step=1)

years_slider = pn.widgets.IntSlider(name='Year', start=2021, end=2022, step=1)

    
yaxis_dimentions = pn.widgets.RadioButtonGroup(
    name='Dimentions', 
    options=['shares','comments','reactions','like','love','haha','wow','sad','angry'],
    button_type='success'
)
```

```python
ipipeline = (
    idf[
        (idf.created_month == months_slider) & 
        (idf.created_year == years_slider)
    ]
    .groupby(['company','created_date'])[yaxis_dimentions].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='created_date')  
    .reset_index(drop=True)
)
ipipeline
```

```python
itable = ipipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size=10, theme=theme)
itable
```

```python
ihvplot = ipipeline.hvplot(x='created_date', y=yaxis_dimentions, by='company', color=PALETTE, line_width=3, height=400, title="Monthly Interaction")
ihvplot
```

```python
depth_hist = ipipeline.hvplot(x=['shares','comments','reactions','like','love','haha','wow','sad','angry'],kind='hist', responsive=True, min_height=200)
```
