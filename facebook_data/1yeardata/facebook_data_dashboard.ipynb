```python
import pandas as pd
import numpy as np
import panel as pn
import datetime as dt
pn.extension('tabulator')

import hvplot.pandas
```

```python
files = ['PrimeLandData1Year.xlsx','BlueOceanGroup﻿Data1Year.xlsx', 'CBHLandsData1Year.xlsx', 'HomeLandsData1Year.xlsx','HouseAndLandsLKData1Year.xlsx', 'SkyLineRealEstateData1Year.xlsx']
```

```python
df = pd.DataFrame()
for file in files:
     if file.endswith('.xlsx'):
         df = df.append(pd.read_excel(file, sheet_name="posts"), ignore_index=True) 
df.head()
```

```python
# Fill NAs with 0s and create GDP per capita column
df = df.fillna(0)
```

```python
# Radio buttons for CO2 measures
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['Shares', 'Comments','Reactions', 'Like', 'Love', 'Haha', 'Angry', 'Wow', 'Sad'] ,
    button_type='success'
)
```

```python
date_range = pn.widgets.DateRangeSlider(
    name='Date Range',
    start=dt.datetime(2021, 7, 1, 00, 00, 00), end=dt.datetime(2022, 8, 1, 23, 59, 59),
    value=(dt.datetime(2021, 7, 1, 00, 00, 00), dt.datetime(2022, 1, 1, 00, 00, 00)),
)

date_range
```

```python
companies = ['Prime Lands', 'Blue Ocean','CBH Lands', 'Home Lands', 'House and Lands', 'Sky Line']

co2_pipeline = (
    df[
        ((dt.datetime.strftime(df.created_time,'%Y-%m-%dT%H:%M:%S')) >= date_range.start) &
        (df.company.isin(companies))
    ]
    .groupby(['company', 'created_time'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='created_time')  
    .reset_index(drop=True)
)
```
