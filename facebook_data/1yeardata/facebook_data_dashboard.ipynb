```python
import pandas as pd
import numpy as np
import panel as pn
import datetime as dt
pn.extension('tabulator')

import hvplot.pandas
```

```python
files = ['PrimeLandData1Year.xlsx','BlueOceanGroupÔªøData1Year.xlsx', 'CBHLandsData1Year.xlsx', 'HomeLandsData1Year.xlsx','HouseAndLandsLKData1Year.xlsx', 'SkyLineRealEstateData1Year.xlsx']
```

```python
df = pd.DataFrame()
for file in files:
     if file.endswith('.xlsx'):
         df = df.append(pd.read_excel(file, sheet_name="posts"), ignore_index=True) 
df.head()
```

```python
df.columns
```

```python
df[df['company'] == 'Prime Lands']
```

```python
df['post_id'] = df['post_id'].replace('', np.nan)
df = df.dropna(axis=0, subset=['post_id'])

df['created_time'] = df['created_time'].replace('', np.nan)
df = df.dropna(axis=0, subset=['created_time'])

df['updated_time'] = df['updated_time'].replace('', np.nan)
df = df.dropna(axis=0, subset=['updated_time'])

df
```

```python
# Fill NAs with 0s and create GDP per capita column
df = df.fillna(0)
```

```python
df['shares_count'] = df['shares_count'].replace(np.nan, 0)
df['comments_count'] = df['comments_count'].replace(np.nan, 0)
df['reactions_count'] = df['reactions_count'].replace(np.nan, 0)
df['like_count'] = df['like_count'].replace(np.nan, 0)
df['love_count'] = df['love_count'].replace(np.nan, 0)
df['haha_count'] = df['haha_count'].replace(np.nan, 0)
df['wow_count'] = df['wow_count'].replace(np.nan, 0)
df['sad_count'] = df['sad_count'].replace(np.nan, 0)
df['angry_count'] = df['angry_count'].replace(np.nan, 0)

df['created_time'] = pd.to_datetime(df['created_time'])
df['updated_time'] = pd.to_datetime(df['updated_time'])

df.sort_values(by='created_time', inplace=True)

df
```

```python
# Make DataFrame Pipeline Interactive
idf = df.interactive()
```

```python
date_range = pn.widgets.DateRangeSlider(
    name='Date Range',
    start=dt.datetime(2021, 7, 1, 00, 00, 00), end=dt.datetime(2021, 8, 1, 00, 00, 00),
    value=(dt.datetime(2021, 7, 2, 00, 00, 00), dt.datetime(2021, 7, 4, 00, 00, 00)),
    format="%x, %X"
)

date_range
```

```python
date_range.value
```

```python

data = {'name':['A', 'B', 'C', 'D'],
        'open_time': [dt.datetime(2021, 7, 1, 21, 0), dt.datetime(2021, 7, 2, 11, 0), dt.datetime(2021, 7, 3, 3, 0),dt.datetime(2021, 7, 3, 2, 0) ],
        'count':[27, 24, 22, 32],
        'size':[34, 56,67,78]
       }

df_sample = pd.DataFrame(data)
df_sample
```

```python
yaxis_sample = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['count', 'size'],
    button_type='success'
)
```

```python
names = ['A', 'B','C', 'D']

sample_pipeline = (
    df[
        (df_sample.open_time >= date_range.start) & (df_sample.open_time <= date_range.end) &
        (df_sample.name.isin(names))
    ]
    .groupby(['name', 'open_time'])[yaxis_sample].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='open_time')  
    .reset_index(drop=True)
)
```
